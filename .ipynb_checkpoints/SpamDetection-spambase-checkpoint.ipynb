```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
import os
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(os.path.join('data', 'spambase', 'spambase.data'), header=None)
df
```

```python
X = df.iloc[:, :48]
Y = df.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size=0.8)
```

```python
nb = MultinomialNB()
nb.fit(X_train, y_train)
print('Multinomial Naive Bayes Classification score ', nb.score(X_test, y_test))
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
print('Decision Tree Classification score ', dt.score(X_test, y_test))
```
