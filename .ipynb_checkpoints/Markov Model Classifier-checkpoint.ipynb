## Label Representation
- Edgar Allan Poe : 0
- Robert Frost : 1

```python
import os
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import pad_sequences
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import pandas as pd
import string
```

## Tokenization and Splitting

```python
MAX_VOCAB_SIZE = 3000
tokenizer = Tokenizer(num_words=MAX_VOCAB_SIZE, oov_token="<OOV>")

with open(os.path.join('data', 'edgar_allan_poe.txt')) as f1 :
    sentences1 = []
    for line in f1 :
        line = line.rstrip().lower()
        if line:
            line = line.translate(str.maketrans('', '', string.punctuation))
            sentences1.append(line)

with open(os.path.join('data', 'robert_frost.txt')) as f2 :
    sentences2 = []
    for line in f2 :
        line = line.rstrip().lower()
        if line:
            line = line.translate(str.maketrans('', '', string.punctuation))
            sentences2.append(line)
    
sentences = np.concatenate((sentences1, sentences2))
Y = np.array([[0] * len(sentences1) + [1] * len(sentences2)]).reshape(-1)
X_train, X_test, y_train, y_test = train_test_split(sentences, Y, train_size=0.8)
del sentences, sentences1, sentences2

tokenizer.fit_on_texts(X_train)
X_train = tokenizer.texts_to_sequences(X_train)
X_test = tokenizer.texts_to_sequences(X_test)

print(len(X_train), len(X_test), Y.shape, sep = '\n')
```

```python
M = len(tokenizer.word_index)
M
```

```python
pi_estimate0 = pd.Series(0, index=tokenizer.word_index.values())
A_estimate0 = pd.DataFrame(0, columns=tokenizer.word_index.values(), index = tokenizer.word_index.values())

pi_estimate1 = pd.Series(0, index=tokenizer.word_index.values())
A_estimate1 = pd.DataFrame(0, columns=tokenizer.word_index.values(), index = tokenizer.word_index.values())

for i in range(len(X_train)):
    sequence = X_train[i]
    label = y_train[i]
    
    if label == 0 :
        pi_estimate0.loc[sequence[0]] += 1
    else :
        pi_estimate1.loc[sequence[0]] += 1
        
    for j in range(len(sequence)-1) :
        word = sequence[j]
        next_word = sequence[j+1]
        if label == 0 :
            A_estimate0.loc[word, next_word] += 1
        else :
            A_estimate1.loc[word, next_word] += 1
```

## Add one Smoothing

```python
A_estimate0 = (A_estimate0 + 1) / (A_estimate0.sum(axis = 1) + M)
pi_estimate0 = (pi_estimate0 + 1) / ((y_train == 0).sum() + M)

A_estimate1 = (A_estimate1 + 1) / (A_estimate1.sum(axis = 1) + M)
pi_estimate1 = (pi_estimate1 + 1) / ((y_train == 1).sum() + M)

display(A_estimate0, pi_estimate0, '\n', A_estimate1, pi_estimate1)
```

```python
log_A0 = np.log(A_estimate0)
log_pi0 = np.log(pi_estimate0)

log_A1 = np.log(A_estimate1)
log_pi1 = np.log(pi_estimate1)
```

## Computing the prior and log prior

```python
p0 = (y_train == 0).sum() / len(y_train)
p1 = (y_train == 1).sum() / len(y_train)
log_p0 = np.log(p0)
log_p1 = np.log(p1)
print('Prior distribution for author Edgar Allan Poe : ' ,p0)
print('Prior distribution for author Robert Frost : ' ,p1)
print('Log Prior distribution for author Edgar Allan Poe : ' ,log_p0)
print('Log Prior distribution for author Robert Frost : ' ,log_p1)
```

```python
def compute_log_likelihood(sequence, log_pi, log_A) :
    log_likelihood = 0
    first_state = sequence[0]
    log_likelihood += log_pi.loc[first_state]
    for j in range(1, len(sequence) - 1) :
        second_state = sequence[j]
        third_state = sequence[j+1]
        log_likelihood += log_A.loc[second_state, third_state]
    return log_likelihood

def predict(inputs, log_pis, log_As, log_priors) :
    predictions = np.zeros(len(inputs))
    for i, input_ in enumerate(inputs):
        posteriors = [(compute_log_likelihood(input_, log_pis[class_], log_As[class_]) + log_priors[class_]) for class_ in range(len(log_pis))]
        k = np.argmax(posteriors)
        predictions[i] = k
    return predictions
```

## Predicting

```python
prediction_X_train = predict(X_train, [log_pi0, log_pi1], [log_A0, log_A1], [log_p0, log_p1])
prediction_X_test = predict(X_test, [log_pi0, log_pi1], [log_A0, log_A1], [log_p0, log_p1])
```

```python
print('Train set accuracy : ', accuracy_score(y_train, prediction_X_train))
print('Train set f1-score : ', f1_score(y_train, prediction_X_train))
print('Train set Confusion Matrix : \n', confusion_matrix(y_train, prediction_X_train))
```

```python
print('Test set accuracy : ', accuracy_score(y_test, prediction_X_test))
print('Test set f1-score : ', f1_score(y_test, prediction_X_test))
print('Test set Confusion Matrix : \n', confusion_matrix(y_test, prediction_X_test))
```

## ROC Curve

```python
fpr_train, tpr_train, thresholds = roc_curve(y_train, prediction_X_train)
fpr_test, tpr_test, thresholds = roc_curve(y_test, prediction_X_test)

# Calculate the AUC (Area Under the ROC Curve)
roc_auc_train = roc_auc_score(y_train, prediction_X_train)
roc_auc_test = roc_auc_score(y_test, prediction_X_test)

# Plot ROC curve
plt.figure(figsize=(5, 3))
plt.plot(fpr_train, tpr_train, color='darkorange', lw=2, label='Train : ROC curve (area = {:.2f})'.format(roc_auc_train))
plt.plot(fpr_test, tpr_test, color='red', lw=2, label='Test : ROC curve (area = {:.2f})'.format(roc_auc_test))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.grid()
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
